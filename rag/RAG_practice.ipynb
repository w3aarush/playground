Building a Retrieval Augmented Generation (RAG) system typically involves several key steps. Here's a common workflow:

* Data Ingestion and Loading: First, you need to load your source data. This could be from various formats like PDFs, text files, databases, or web pages. Libraries like LangChain offer document loaders for many different sources.

* Text Splitting: Large documents need to be broken down into smaller, manageable chunks. This is crucial because LLMs have token limits, and it also allows for more granular retrieval. During this step, you often introduce overlapping between chunks to maintain context, as we discussed.

* Embedding Generation: Each text chunk is then converted into a numerical vector representation, called an embedding. These embeddings capture the semantic meaning of the text. HuggingFaceEmbeddings, which you've imported, is a common choice for this.

* Vector Store Indexing: The generated embeddings, along with their corresponding text chunks, are stored in a vector database or vector store. FAISS, which you've also imported, is a popular choice for efficiently storing and querying these embeddings. This step creates an index that allows for fast similarity searches.

* Retrieval: When a user asks a question, that query is also converted into an embedding. This query embedding is then used to search the vector store for the most semantically similar text chunks from your knowledge base. These retrieved chunks are the 'context' for the LLM.

* Augmented Generation: Finally, the retrieved text chunks (the context) and the user's original query are passed to a Large Language Model (LLM). The LLM then uses this provided context to generate a more accurate, informed, and relevant answer than it would be able to produce based on its pre-trained knowledge alone.

In summary, the RAG workflow involves getting your data, breaking it into searchable pieces, converting those pieces into a numerical format, storing them efficiently, retrieving the most relevant pieces for a given query, and then using those pieces to help an LLM generate a better response.
```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.prompts import PromptTemplate
from langchain_community.document_loaders import PyPDFLoader
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=50)
```

```python
docs = splitter.create_documents([text])
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
retriver = vectorstore.as_retriever(search_kwargs={'k':3})
```

```python
def format_docs(docs):
    return "\n\n".join(d.page_content for d in docs)
```

```python
prompt = PromptTemplate(
    input_variable = ['context', 'question'],
    template = """
    Use only the context below to answer the question.
    context:
    {context}

    question: {question}

    Answer :
    """
)
```

```python
llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")
```

```python
def rag_answer(query):
    r_docs = retriver.invoke(query)
    context = format_docs(r_docs)
    full_prompt = prompt.format(context=context, question = query)
    response = llm.invoke(full_prompt)

    return response.content
```

```python
query = "What is python used for?"

answer = rag_answer(query)
```
